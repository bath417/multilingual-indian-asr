# Multilingual Indian Automatic Speech Recognition (ASR) System
This project implements a multilingual speech recognition system for Indian languages such as Telugu and Tamil. The system converts speech to text using the Whisper ASR model and translates the recognized text into multiple languages.

## Problem Statement
India has many languages, and speech recognition systems must support multiple languages. Many Indian languages are low-resource languages with limited datasets. The goal of this project is to build a multilingual ASR system that can recognize speech in Indian languages and translate it into multiple languages.

## Testing and Results
The system was tested using Telugu and Tamil speech inputs. The model successfully recognized the speech and translated the output into multiple languages.

```python
import gradio as gr
from deep_translator import GoogleTranslator
import whisper

asr_model = whisper.load_model("large")

def speech_to_multilingual(audio):

    result = asr_model.transcribe(audio, language="te", task="transcribe")
    recognized_text = result["text"]

    english = GoogleTranslator(source='auto', target='en').translate(recognized_text)
    hindi = GoogleTranslator(source='auto', target='hi').translate(recognized_text)
    tamil = GoogleTranslator(source='auto', target='ta').translate(recognized_text)
    telugu = GoogleTranslator(source='auto', target='te').translate(recognized_text)

    return f"""
Recognized Text:
{recognized_text}

English:
{english}

Hindi:
{hindi}

Tamil:
{tamil}

Telugu:
{telugu}
"""

interface = gr.Interface(
    fn=speech_to_multilingual,
    inputs=gr.Audio(type="filepath"),
    outputs="text",
    title="Multilingual Indian ASR System",
    description="Upload Telugu or Tamil speech to convert speech into multilingual text"
)

interface.launch()
```
